### Loading Processed dataframes

```python
from owlready2 import *
import pandas as pd
df1 = pd.read_csv('Characters.csv', error_bad_lines=False, encoding='latin-1')
import csv

df1['Skills'] = df1['Skills'].str.replace("|", "_")
df1['Job'] = df1['Job'].str.replace("|", "_")
df1['HairColor'] = df1['HairColor'].str.replace("|", "_")
df1["BloodStatus"] = df1["BloodStatus"].str.replace("-", "_")

df2  = pd.read_csv("Spells.csv")
df2 = df2.drop(['SpellID'],axis=1)
```

## Loading Ontology

```python
onto = get_ontology ("MyOnto.owl")
#onto = get_ontology("http://www.semanticweb.org/ana/ontologies/2022/11/HarryPotterMovies").load()
try:
  onto.import_from_owl()
except Exception as e:
  print(e)
onto.load()
print(onto)
```

### POPULATING

#### Assigning Gender

```python
# set the gender
with onto:
    for i in df1['Character']:             
        char = onto.Character(i)        
        if (df1[df1['Character'] == i]['Gender'].iloc[0] == "Female"):            
            char.hasGender = "Female"
        elif (df1[df1['Character'] == i]['Gender'].iloc[0] == "Male"):
            char.hasGender = "Male"        
```

#### Character haswand Wand

```python
# has wand
with onto:
    for i in df1['Character']:
        for j in df1['Wand']:     
            if (df1[df1['Character'] == i]['Wand'].iloc[0] == j): 
                onto.Character(i).hasWand = onto.Wand(j)
#cheching 
print(onto.PeterPettigrew.hasWand)
```

#### Character hasJob Job

```python
with onto:
    for i in df1['Character']:
        for j in df1['Job']:
       #onto.harryPotter.hasJob = onto.student 
            if (df1[df1['Character'] == i]['Job'].iloc[0] == j): 
                onto.Character(i).hasJob = onto.Job(j)

#checking 
print (onto.HarryJamesPotter.hasJob)
print(onto.RemusJohnLupin.hasJob)
print(onto.PeterPettigrew.hasJob)               
```

#### Character hasPatronus Patronus

```python
Patronus_instances =[]
with onto:
    for i in df1['Character']:
        for j in df1['Patronus']:     
            if (df1[df1['Character'] == i]['Patronus'].iloc[0] == j): 
                onto.Character(i).hasPatronus = onto.Patronus (j)
        Patronus_instances.append(onto.Character(i).hasPatronus)
        
print(onto.HarryJamesPotter.hasPatronus)
```

#### Character hasSkills Skills

```python
with onto:
    for i in df1['Character']:
        for j in df1['Skills']:     
            if (df1[df1['Character'] == i]['Skills'].iloc[0] == j): 
                onto.Character(i).hasSkills = onto.Skills (j)
        
print(onto.HarryJamesPotter.hasSkills)
```

#### Character hasHairColor, hasEyeColor

```python
with onto:
    for i in df1['Character']:
        for j in df1['HairColor']:     
            if (df1[df1['Character'] == i]['HairColor'].iloc[0] == j): 
                onto.Character(i).hasHairColor = onto.HairColor (j)
        
print(onto.PeterPettigrew.hasHairColor)

# eye color
with onto:
    for i in df1['Character']:
        for j in df1['EyeColor']:     
            if (df1[df1['Character'] == i]['EyeColor'].iloc[0] == j): 
                onto.Character(i).hasEyeColor = onto.EyeColor (j)
print(onto.PeterPettigrew.hasEyeColor)
```

#### Character BelongsToSpecies Species

```python
# belongs to species

with onto:
    for i in df1['Character']:
        char = onto.Character(i)
        species = df1[df1['Character'] == i]['Species'].iloc[0]
        species_class = None
        try:
            species_class = onto.search_one(iri='*'+species)
        except Exception:
            pass
        if species_class:
            char.BelongsToSpecies = species_class

print(onto.MyrtleElizabethWarren.BelongsToSpecies)
print(onto.RemusJohnLupin.BelongsToSpecies)
```

#### Character BelongsToHouse House

```python
# Belongs to house
with onto:
    for i in df1['Character']:
        for j in df1['House']:
       #onto.harryPotter.hasJob = onto.student 
            if (df1[df1['Character'] == i]['House'].iloc[0] == j): 
                onto.Character(i).BelongsToHouse = onto.House(j)


            
print(onto.MyrtleElizabethWarren.BelongsToHouse)
print(onto.RemusJohnLupin.BelongsToHouse)
```

#### Character hasBloodStatus

```python
with onto:
    for i in df1['Character']:
        char = onto.Character(i)
        blood_status = df1[df1['Character'] == i]['BloodStatus'].iloc[0]
        char.hasBloodStatus = onto[blood_status]

    
print(onto.MyrtleElizabethWarren.hasBloodStatus)
```
